## K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

Creating an instance of a K Means model with 2 clusters. Beacuse we already know our clusters.(Private and Public)

```python
kmc = KMeans(n_clusters=2)
```

I will fit the model to all the data except for the Private label.

```python
kmc.fit(df.drop('Private',axis=1))
```

Here are the cluster center vectors.

```python
kmc.cluster_centers_
```

## Evaluation
Creating a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
df['Cluster'] = df['Private'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
df.tail()
```

Creating a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(df['Cluster'],kmc.labels_))
print(classification_report(df['Cluster'],kmc.labels_))
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(df['F.Undergrad'],df['Outstate'],c=kmc.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(df['F.Undergrad'],df['Outstate'],c=df['Cluster'],cmap='rainbow')
```
